```python
data2017[data2017['CASE_STATUS']=='CERTIFIED'].groupby('SOC_NAME').size()
```

```python
certified_by_occ=data2017[data2017['CASE_STATUS']=='CERTIFIED'].groupby('SOC_NAME').size().sort_values(ascending=False)[:10].reset_index(name='count')
certified_by_occ.head()
```

```python
import math
```

```python
# What proportion of certifications does this represent?
def certifications_by_occ(x):
    #return  '{:.1%}'.format((x /float(certified_by_occ['count'].sum()) * 100), .0)
    result=round((x /float(certified_by_occ['count'].sum()) * 100), 1)
    return "{0:.1f}%".format(result)
certified_by_occ['percentage'] = certified_by_occ['count'].apply(certifications_by_occ)

```

```python
certified_by_occ=certified_by_occ.sort_values(['count', 'SOC_NAME'], ascending=[False, True])
certified_by_occ.head()
```

```python
myformatter = "; "
certified_by_occ.to_string(header=True, index=False,  index_names=False)
```
