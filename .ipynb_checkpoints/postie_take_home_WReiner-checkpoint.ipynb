#### Scenario:

It is July 3rd, 2017. A data analyst just started at your company on July 1st, and has done an analysis to figure out the sales for 2017-07-03 were $164065.00.

#### Technical Details:

Each file contains rows with the following data:

* timestamp - the date of the transaction
* website_id - the integer value of the website the transaction took place on
* customer_id - the integer identifier for the customer
* app_version - the version of the software
* placeholder - a placeholder column
* checkout_amount - the amount in dollars of the complete transaction
* url - the full website url that instantiated the transaction, with product names and product count params

#### Analysis
#### 1. The analyst wants you to figure out why their 2017-07-03 sales value is much lower than their previous day sales summary.*

There are a combination of factors that contributed to the lower total sales value on 7-03:
1.  Fewer purchases were made in the early hours of 07-03. This is seems to be to the release of app_version 1.2, as it does not occur in the data from previous days. Customers who were updating the app could not make purchases during these hours.
2.  Customers were, at times, charged $0.00 for Hazelnuts, although this occurred on days other than July 3rd.
3.  Customers were charged negative amounts (or made returns) on July 3rd only, This definitely contributed to the lower total sales.

#### 2. The new analyst would also like a report detailing other key metrics for the system. Other than the average sales value per day, are there any other metrics that they should be using? 

Other than average sales value per day, there are a number of other metrics that could be useful to the client. These include:

* total sales per day
* distribution of sales by hour within each day
* average frequency of sales per hour
* weekly sales distribution (This cannot be assessed from the available data, but it would be a useful metric.)

Knowing which days and which time-windows have the highest purchase activity would enable the client to release new products, offer deals, and maximize earnings from advertising by charging advertisers more on these days/during these times. It could also help the client determine how many customer service representatives to have available. Optimizing the could lead to higher profits during busy times when customers need help completing their purchases, and lead to saving money by having fewer representatives during slower times for business. As previously mentioned, these metrics could also help the client determine when releasing a new version of the app. would have the least affect on sales.

Other metrics include:

* Min, max, and average number of transactions per customer, so that customers with the highest lifetime values can be profiled and special programs or deals can be offered to them. This could also help the client to identify which customers would have the highest ROI for direct mailing efforts.

* All of these metrics should also be calculated per website if the client would like to compare each site's profitability.


In this scenario, average sales value is okay to use, but it does not give a complete picture, especially due to the unique bugs in the app. It would be more informative to use additional key metrics I have listed above, as well as others, like average sales per hour for each item.



#### 3. What information can you extract from the urls?

* From the urls you can extract the domain that the customer visited, what items they purchased, and the quentities of each item. Therefore you can also infer product prices from these data because each item was the only item purchased by at least one customer.  

Is there any other information that you believe would be useful to understand what is going on?

This way, new versions of the app (barring emergency updates if security issues arise, for example) can be released at these times and have the least effect on profits.
